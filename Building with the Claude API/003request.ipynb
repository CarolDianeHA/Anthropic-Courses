```python
# Load env variables

from dotenv import load_dotenv
load_dotenv()

# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }

    if system:
        params["system"] = system

    message = client.messages.create(
        **params
    )
    return message.content[0].text
```

```python
messages = []

add_user_message(messages, "Generate a one sentence movie idea.")

response = chat(messages, temperature=1.0)

print(response)

# Here's a movie idea:
# temperature=0.0 A retired safecracker with dementia must rely on the muscle memory of his criminal past to recover a stolen item that holds the only clue to his forgotten identity.
# temperature=1.0 A washed-up lighthouse keeper discovers that the mysterious light he's been guiding to shore every night belongs to a ghost ship reliving its final tragic voyage, and he must uncover the truth behind its sinking to finally set the crew free.
```
